**¿Cómo es la proporción de géneros y grupos etarios que han estado involucrados en este tipo de delito? ¿Han variado con el paso de los años?**


```python
percent_ge_vf = vf.groupby('GRUPO ETARIO').agg({'CANTIDAD': 'sum'})
total_casos_vf = percent_ge_vf['CANTIDAD'].sum()                                 
percent_ge_vf = percent_ge_vf.groupby('GRUPO ETARIO').agg({'CANTIDAD': lambda x:100 * x / total_casos_vf})
percent_ge_vf.head()
```

```python
# gráfico de barras de frecuencias relativas.
plot = (percent_ge_vf).plot(
kind='bar', title='% GRUPO ETARIO - Violencia Familiar')
```

**De acuerdo al analisis, el porcentaje en casos de violencia familiar para grupo etario es 90.8% para Adultos, y 4% para Adolescentes y Menores**

```python
percent_g_vf = vf.groupby('GENERO').agg({'CANTIDAD': 'sum'})
total_casos_vf = percent_g_vf['CANTIDAD'].sum()                                 
percent_g_vf = percent_g_vf.groupby('GENERO').agg({'CANTIDAD': lambda x:100 * x / total_casos_vf})
percent_g_vf.head()
```

```python
# gráfico de barras de frecuencias relativas.
plot = (percent_g_vf).plot(
kind='bar', title='% Genero - Violencia Familiar')
```
